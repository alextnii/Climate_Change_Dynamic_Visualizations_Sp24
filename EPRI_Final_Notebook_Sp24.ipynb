```python
def create_decade(df, in_place = True):
    year_series = df['time'].dt.year
    if in_place == True:
        df['decade'] = (year_series // 10 * 10)
    else:
        df_copy = df.copy(deep = True)
        df_copy['decade'] = (year_series // 10 * 10)
        return df_copy

def create_decade_widget():
    decade_series = create_decade_series(tmax_df)
    all_decades_unique = list(decade_series.unique())
    decade_widget = pn.widgets.Select(name = 'Select Decade', options = all_decades_unique)
    return decade_widget
```

```python
# Correct plot:

color_dictionary = {'tmin': 'blue', 'tmax': 'red', 'precip': 'green'}

# a function that plots a general one that can take in the specific df and y limits
def create_plot(df, variable, city, model, scen, dec, ylim):
    filtered_df = df[(df['model'] == model) & (df['City'] == city) &
                     (df['scenario'] == scen) & (df['decade'] == dec)]
    filtered_df['month'] = filtered_df['time'].dt.month
    filtered_df['time'] = filtered_df['time'].astype(str)

    mean_df = filtered_df.groupby('month')[variable].agg('mean').reset_index().rename(columns={variable: 'Mean'})
    sd_df = filtered_df.groupby('month')[variable].agg('std').reset_index().rename(columns={variable: 'SD'})
    if variable == 'tmax':
        max_df = filtered_df.groupby('month')[variable].agg('max').reset_index().rename(columns={variable: 'Max'})
        filtered_df = mean_df.merge(sd_df, on='month').merge(max_df, on='month')
    elif variable == 'tmin':
        min_df = filtered_df.groupby('month')[variable].agg('min').reset_index().rename(columns={variable: 'Min'})
        filtered_df = mean_df.merge(sd_df, on='month').merge(min_df, on='month')
    else:
        filtered_df = mean_df.merge(sd_df, on='month')
    filtered_df['Mean - 1SD'] = filtered_df['Mean'] - filtered_df['SD']
    filtered_df['Mean + 1SD'] = filtered_df['Mean'] + filtered_df['SD']

    hover_area = HoverTool(tooltips=[("month", "$x"), ("Precip / Temp", "$y")])
    area_plot = filtered_df.hvplot.area(x='month', y='Mean - 1SD', y2='Mean + 1SD', alpha=0.3, line_width=0.8,
                                        height=350, width=1500, color=color_dictionary[variable],
                                        ylim=ylim, label= '+/- 1SD for ' + variable.capitalize()).opts(tools=[hover_area], xticks=list(range(1, 13)))

    line_plot = filtered_df.hvplot.line(x='month', y='Mean', line_width=0.8,
                                        height=350, width=1500, color=color_dictionary[variable],
                                        ylim=ylim, label=variable.capitalize() + ' Mean').opts(tools=['save', 'pan', 'wheel_zoom', 'box_zoom', 'reset'])
    if variable == 'tmax':
        dotted = hv.Curve(filtered_df, 'month', 'Max', label='Max').opts(line_dash='dashed', tools=[hover_area])
        area_plot.opts(tools = [hover_area])
        line_plot.opts(tools=['save', 'pan', 'wheel_zoom', 'box_zoom', 'reset'], shared_axes = False)
        overlay_plot = area_plot * line_plot * dotted 
    elif variable == 'tmin':
        dotted = hv.Curve(filtered_df, 'month', 'Min', label='Min').opts(color='blue', line_dash='dashed', tools=[hover_area])
        area_plot.opts(tools = [hover_area])
        line_plot.opts(tools=['save', 'pan', 'wheel_zoom', 'box_zoom', 'reset'], shared_axes = False)
        overlay_plot = area_plot * line_plot * dotted 
    else:
        area_plot.opts(tools = [hover_area])
        line_plot.opts(tools=['save', 'pan', 'wheel_zoom', 'box_zoom', 'reset'], shared_axes = False)
        overlay_plot = area_plot * line_plot
    return overlay_plot.opts(shared_axes=False)

# Temp plot
@pn.depends(city_widget.param.value, model_widget.param.value, scen_widget.param.value, dec_widget.param.value)
def create_temp_plots(city, model, scen, dec):
    tmax_plot = create_plot(tmax_df, 'tmax', city, model, scen, dec, ylim=(-10, 110))
    tmin_plot = create_plot(tmin_df, 'tmin', city, model, scen, dec, ylim=(-10, 110))
    combined_overlay_plot = tmax_plot * tmin_plot
    return combined_overlay_plot.opts(xlabel='Month', ylabel='Temperature (F)', shared_axes=False)

# Precip plot
@pn.depends(city_widget.param.value, model_widget.param.value, scen_widget.param.value, dec_widget.param.value)
def create_precip_plot(city, model, scen, dec):
    precip_plot = create_plot(precip_df, 'precip', city, model, scen, dec, ylim=(-1, 1))
    return precip_plot.opts(xlabel='Month', ylabel='Precipitation (mm)', shared_axes=False)
```

```python
models_data2 = xr.open_dataset('Model_data2.nc')
m_precip2 = models_data2.precip
precip_df2 = m_precip2.to_dataframe().reset_index()
t_df2 = models_data2.tmax
tmax_df2 = t_df2.to_dataframe().reset_index()
tm_df2 = models_data2.tmin
tmin_df2 = tm_df2.to_dataframe().reset_index()
historical_data2 = xr.open_dataset('Historical_data2.nc')
h_tmax_df2 = historical_data2.tmax.to_dataframe().reset_index()

precip_df2['City'] = precip_df2['Locations'].map(locations_to_cities)
tmax_df2['City'] = tmax_df2['Locations'].map(locations_to_cities)
tmin_df2['City'] = tmin_df2['Locations'].map(locations_to_cities)
```

```python
@pn.depends(city_selector.param.value, scenario_selector.param.value, model_selector.param.value, visualization_selector.param.value, timeperiod_selector.param.value)
def rlevel_rperiod_plot(city, scenario, model, visualization_type, timeperiod):
    if visualization_type == visualization_types[0]:
        constrained_df = precip_df2[(precip_df2['City'] == city) &
                                   (precip_df2['scenario'] == scenario) &
                                   (precip_df2['model'] == model)]
        
        constrained_df = constrained_df[(constrained_df['time'].dt.year >= int(timeperiod[0:4])) & (constrained_df['time'].dt.year <= int(timeperiod[-4:]))]
        constrained_df = constrained_df.copy()
        constrained_df['Year number'] = constrained_df['time'].dt.year - int(timeperiod[0:4])

        grouped_df = constrained_df.groupby('Year number').agg(max).reset_index()
        sorted_df = grouped_df.sort_values("precip", ascending = False)
        sorted_df['return_period'] = range(1, len(sorted_df) + 1)
        n = len(sorted_df)
        sorted_df['return_period'] = (n + 1) / (sorted_df['return_period'])
        
        line_plot = sorted_df.hvplot.scatter(x='return_period', y='precip', line_width=0.8,
                                 height=500, width=500, color='blue',
                                 xlabel='Return Period (Years)', ylabel='Return Level (Precip)', title='Return Level vs Return Period')
        def logarithmic_curve(x, a, b):
            return a * np.log(x) + b

        # Fit the curve to the data
        params, covariance = curve_fit(logarithmic_curve, sorted_df['return_period'], sorted_df['precip'])

        # Extract the fitted parameters
        a, b = params

        # Generate the best fit line
        x_fit = np.linspace(min(sorted_df['return_period']), max(sorted_df['return_period']), 100)
        y_fit = logarithmic_curve(x_fit, a, b)

        fit_plot = hv.Curve((x_fit, y_fit), label='Best Fit Line')
        
        return line_plot * fit_plot
        return line_plot * confidence_bands * fit_plot

    
    elif visualization_type == visualization_types[1]:
        constrained_df = tmax_df2[(tmax_df2['City'] == city) & 
                                 (tmax_df2['scenario'] == scenario) & 
                                 (tmax_df2['model'] == model)]
        constrained_df = constrained_df[(constrained_df['time'].dt.year >= int(timeperiod[0:4])) & (constrained_df['time'].dt.year <= int(timeperiod[-4:])) & (constrained_df['time'].dt.month == 6)]
        constrained_df = constrained_df.copy()
        constrained_df['Year number'] = constrained_df['time'].dt.year - int(timeperiod[0:4])
        
        #group by year and add column days over 90 (num days in june w tmax over 90)
        def daysOver90(series):
            return (series > 90).sum()
        constrained_df['daysOver90'] = constrained_df.groupby('Year number')['tmax'].transform(daysOver90)
        grouped_df = constrained_df.drop_duplicates(subset='Year number')

        sorted_df = grouped_df.sort_values("daysOver90", ascending = False)
        sorted_df['return_period'] = range(1, len(sorted_df) + 1)
        n = len(sorted_df)
        sorted_df['return_period'] = (n + 1) / (sorted_df['return_period'])
        line_plot = sorted_df.hvplot.scatter(x='return_period', y='daysOver90', line_width=0.8,
                                 height=500, width=500, color='blue',
                                 xlabel='Return Period (Years)', ylabel='Return Level (TMax)', title='Return Level vs Return Period')
        def logarithmic_curve(x, a, b):
            return a * np.log(x) + b

        # Fit the curve to the data
        params, covariance = curve_fit(logarithmic_curve, sorted_df['return_period'], sorted_df['daysOver90'])

        # Extract the fitted parameters
        a, b = params

        # Generate the best fit line
        x_fit = np.linspace(min(sorted_df['return_period']), max(sorted_df['return_period']), 100)
        y_fit = logarithmic_curve(x_fit, a, b)

        fit_plot = hv.Curve((x_fit, y_fit), label='Best Fit Line')
        
        return line_plot * fit_plot
        return line_plot * fit_plot * confidence_bands
    
    elif visualization_type == visualization_types[2]:
        constrained_df = tmax_df2[(tmax_df2['City'] == city) & 
                                 (tmax_df2['scenario'] == scenario) & 
                                 (tmax_df2['model'] == model)]
        constrained_df = constrained_df[(constrained_df['time'].dt.year >= int(timeperiod[0:4])) & (constrained_df['time'].dt.year <= int(timeperiod[-4:])) & (constrained_df['time'].dt.month == 12)]
        constrained_df = constrained_df.copy()
        constrained_df['Year number'] = constrained_df['time'].dt.year - int(timeperiod[0:4])
        def daysUnder32(series):
            return (series < 32).sum()
        constrained_df['daysUnder32'] = constrained_df.groupby('Year number')['tmax'].transform(daysUnder32)
        grouped_df = constrained_df.drop_duplicates(subset='Year number')
        
        sorted_df = grouped_df.sort_values("daysUnder32", ascending = False)
        sorted_df['return_period'] = range(1, len(sorted_df) + 1)
        n = len(sorted_df)
        sorted_df['return_period'] = (n + 1) / (sorted_df['return_period'])
        line_plot = sorted_df.hvplot.scatter(x='return_period', y='daysUnder32', line_width=0.8,
                                 height=500, width=800, color='blue',
                                 xlabel='Return Period (Years)', ylabel='Return Level (TMax)', title='Return Level vs Return Period')
        
        def logarithmic_curve(x, a, b):
            return a * np.log(x) + b

        # Fit the curve to the data
        params, covariance = curve_fit(logarithmic_curve, sorted_df['return_period'], sorted_df['daysUnder32'])

        # Extract the fitted parameters
        a, b = params

        # Generate the best fit line
        x_fit = np.linspace(min(sorted_df['return_period']), max(sorted_df['return_period']), 100)
        y_fit = logarithmic_curve(x_fit, a, b)

        fit_plot = hv.Curve((x_fit, y_fit), label='Best Fit Line')
        
        return line_plot * fit_plot        
        return line_plot * fit_plot * confidence_bands

    
    elif visualization_type == visualization_types[3]:
        constrained_df = tmin_df2[(tmin_df2['City'] == city) & 
                                 (tmin_df2['scenario'] == scenario) & 
                                 (tmin_df2['model'] == model)]
        constrained_df = constrained_df[(constrained_df['time'].dt.year >= int(timeperiod[0:4])) & (constrained_df['time'].dt.year <= int(timeperiod[-4:])) & (constrained_df['time'].dt.month == 6)]
        constrained_df = constrained_df.copy()
        constrained_df['Year number'] = constrained_df['time'].dt.year - int(timeperiod[0:4])
        def daysOver70(series):
            return (series > 70).sum()
        constrained_df['daysOver70'] = constrained_df.groupby('Year number')['tmin'].transform(daysOver70)
        grouped_df = constrained_df.drop_duplicates(subset='Year number')
        
        sorted_df = grouped_df.sort_values("daysOver70", ascending = False)
        sorted_df['return_period'] = range(1, len(sorted_df) + 1)
        n = len(sorted_df)
        sorted_df['return_period'] = (n + 1) / (sorted_df['return_period'])
        line_plot = sorted_df.hvplot.scatter(x='return_period', y='daysOver70', line_width=0.8,
                                 height=500, width=500, color='blue',
                                 xlabel='Return Period (Years)', ylabel='Return Level (TMin)', title='Return Level vs Return Period')
        def logarithmic_curve(x, a, b):
            return a * np.log(x) + b

        # Fit the curve to the data
        params, covariance = curve_fit(logarithmic_curve, sorted_df['return_period'], sorted_df['daysOver70'])

        # Extract the fitted parameters
        a, b = params

        # Generate the best fit line
        x_fit = np.linspace(min(sorted_df['return_period']), max(sorted_df['return_period']), 100)
        y_fit = logarithmic_curve(x_fit, a, b)

        fit_plot = hv.Curve((x_fit, y_fit), label='Best Fit Line')
        
        return line_plot * fit_plot        
        return line_plot * fit_plot * confidence_bands

    elif visualization_type == visualization_types[4]:
        constrained_df = tmin_df2[(tmin_df2['City'] == city) & 
                                 (tmin_df2['scenario'] == scenario) & 
                                 (tmin_df2['model'] == model)]
        constrained_df = constrained_df[(constrained_df['time'].dt.year >= int(timeperiod[0:4])) & (constrained_df['time'].dt.year <= int(timeperiod[-4:])) & (constrained_df['time'].dt.month == 12)]
        constrained_df = constrained_df.copy()
        constrained_df['Year number'] = constrained_df['time'].dt.year - int(timeperiod[0:4])
        
        def daysUnder0(series):
            return (series < 0).sum()
        constrained_df['daysUnder0'] = constrained_df.groupby('Year number')['tmin'].transform(daysUnder0)
        grouped_df = constrained_df.drop_duplicates(subset='Year number')
        
        sorted_df = grouped_df.sort_values("daysUnder0", ascending = False)
        sorted_df['return_period'] = range(1, len(sorted_df) + 1)
        n = len(sorted_df)
        sorted_df['return_period'] = (n + 1) / (sorted_df['return_period'])
        line_plot = sorted_df.hvplot.scatter(x='return_period', y='daysUnder0', line_width=0.8,
                                 height=500, width=500, aspect = 1, color='blue',
                                 xlabel='Return Period (Years)', ylabel='Return Level (TMin)', title='Return Level vs Return Period')
        def logarithmic_curve(x, a, b):
            return a * np.log(x) + b

        # Fit the curve to the data
        params, covariance = curve_fit(logarithmic_curve, sorted_df['return_period'], sorted_df['daysUnder0'])

        # Extract the fitted parameters
        a, b = params

        # Generate the best fit line
        x_fit = np.linspace(min(sorted_df['return_period']), max(sorted_df['return_period']), 100)
        y_fit = logarithmic_curve(x_fit, a, b)

        fit_plot = hv.Curve((x_fit, y_fit), label='Best Fit Line')
 
        return line_plot * fit_plot
        return line_plot * fit_plot * confidence_bands
```
